```python
#LAB - 02
#Create a dataset of Healthcare and perform operations on it.

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
df = pd.DataFrame({"Patient_ID":[101,102,103,104,105],
                  "Patient_Name":["Harhsit","Neha","Raj","Preeti","Rahul"],
                  "Age":[25,28,38,45,50],
                  "Disease":["Fever","Dengue","Malaria","Eye Flu","Typhoid"],
                  "Income":[30000,35000,42000,25000,50000],
                  "Gender":["M","F","M","F","M"],
                  "BP":[120,130,140,150,165],
                  "Weight":[75,70,80,72,66]})
print(df)
```

```python
df.isnull()
```

```python
x = df["Income"]
y = df["BP"]
plt.xlabel("Income")
plt.ylabel("BP")
plt.scatter(x,y)
plt.grid(True)
plt.show()
```

```python
plt.hist(df["Age"],bins = 5)
plt.xlabel("Age")
plt.ylabel("Frequency")
plt.title("Age Distribution")
plt.show()
```

```python
df["Disease"].value_counts().plot.pie(autopct = "%0.1f%%")
plt.title("Disease Distribution")
plt.ylabel("")
plt.show()
```

```python
plt.boxplot(df["BP"])
plt.ylabel("Blood Pressure")
plt.title("BP BoxpLot")
plt.show()
```

```python
plt.bar(df["Patient_Name"],df["Income"])
plt.xlabel("Patient_Name")
plt.ylabel("Income")
plt.title("Income of Patients")
plt.show()
```

```python
plt.plot(df["Patient_Name"],df["BP"],marker = 'o')
plt.xlabel("Patient_Name")
plt.ylabel("Blood Pressure")
plt.title("Blood Pressure of Patients")
plt.show()
```
