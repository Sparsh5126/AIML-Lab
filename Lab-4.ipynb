```python
df["Model"] = df["Model"].fillna("Hyundai")
df["Manufacturer"] = df["Manufacturer"].fillna("Volvo")
df["Colour"] = df["Colour"].fillna("Purple")
df["Price"] = df["Price"].fillna(df["Price"].mean())
df["Engine Capacity"] = df["Engine Capacity"].fillna(df["Engine Capacity"].mean())
df
```

```python
plt.scatter(df["Model"],df["Price"])
plt.xlabel("Car Model")
plt.ylabel("Car Price")
plt.title("Model vs Price")
plt.grid(True)
plt.show()
```

```python
df["Engine Capacity"].hist(bins = 10)
plt.xlabel("Engine Capacity")
plt.ylabel("Frequency")
plt.title("Car Price Histogram")
plt.show()
```

```python
plt.boxplot(df["Mileage"].dropna())
plt.ylabel("Car Mileage")
plt.title("Boxplot of Car Mileage")
plt.show()
```

```python
df["Mileage"].dropna().value_counts().plot(kind = "bar")
plt.xlabel("Car Mileage")
plt.ylabel("Count")
plt.title("Bar Graph of Car Mileage")
plt.show()
```

```python
plt.plot(df["Colour"],df["Manufacturer"])
plt.xlabel("Car Colour")
plt.ylabel("Car Manufacturer")
plt.title("Line Plot of Colour vs Manufacturer")
plt.show()
```

```python
df["Manufacturer"].value_counts().plot.pie(autopct = "%0.1f%%")
plt.title("Car Manufacturer")
plt.ylabel("")
plt.show()
```
