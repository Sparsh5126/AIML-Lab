```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
df = pd.DataFrame({"Cust_ID":[101,102,103,104,105,106,107],
                   "Cust_Name":["Ayush","Piyush",np.nan,"Rahul","Sam",np.nan,"Raj"],
                   "P_Name":["Facewash","Soap","Toothpaste","Cookies",np.nan,"Cereals","Books"],
                   "P_Cost":[70,60,45,np.nan,50,110,60],
                   "P_Quantity":[2,5,1,5,3,4,np.nan],
                   "Pay_Mode":["Cash","Credit Card","Credit Card",np.nan,"Cash",np.nan,"Cash"]})
print(df)
```

```python
df["Cust_Name"] = df["Cust_Name"].fillna("x")
print(df)
```

```python
plt.scatter(df["P_Cost"], df["P_Quantity"])
plt.xlabel("Product Cost")
plt.ylabel("Product Quantity")
plt.title("Scatter Plot: Product Cost vs Quantity")
plt.grid(True)
plt.show()
```

```python
plt.hist(df["P_Cost"].dropna())
plt.xlabel("Product Cost")
plt.ylabel("Frequency")
plt.title("Histogram of Product Cost")
plt.show()
```

```python
plt.boxplot(df["P_Cost"].dropna())
plt.ylabel("Product Cost")
plt.title("Box Plot of Product Cost")
plt.show()
```

```python
df["Pay_Mode"].value_counts().plot(kind="bar")
plt.xlabel("Payment Mode")
plt.ylabel("Count")
plt.title("Bar Graph of Payment Modes")
plt.show()
```

```python
plt.plot(df["Cust_ID"], df["P_Cost"])
plt.xlabel("Customer ID")
plt.ylabel("Product Cost")
plt.title("Line Plot: Customer ID vs Product Cost")
plt.show()
```
