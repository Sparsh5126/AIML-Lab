```python
#LAB - 01
#Numpy Questions

import numpy as np
arr_1 = np.array([[2,4,6],[2,3,5],[7,8,9]])
arr_2 = np.array([[9,8,7],[6,5,4],[3,2,1]])
print("Matrix Addition:\n",arr_1+arr_2)
print("Matrix Subtraction:\n",arr_1-arr_2)
print("Elementwise multiplication:\n",arr_1*arr_2)
print("Matrix multiplication:\n",arr_1@arr_2)
print("Determinant of arr_1:",np.linalg.det(arr_1))
print("Rank of arr_1:\n",np.linalg.matrix_rank(arr_1))
print("Inverse of arr_1:\n",np.linalg.inv(arr_1))
eigenvalues,eigenvectors = np.linalg.eig(arr_1)
print("Eigen values of arr_1:\n",eigenvalues)
print("Eigen vectors of arr_1:\n",eigenvectors)
print("Maximum of arr_1:\n",np.max(arr_1))
print("Minimum of arr_1:\n",np.min(arr_1))
print("Median of arr_1:\n",np.median(arr_1))
print("Mean of arr_1:\n",np.mean(arr_1))
print("Standard Deviation of arr_1:\n",np.std(arr_1))
```

```python
#Pandas , Matplotlib , Seaborn Questions

import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
df = pd.DataFrame({"Name":["Ayush","Raj","Krish","Bobby","Sam"],
                  "Age":[19,np.nan,18,21,19],
                  "Marks":[85,78,np.nan,88,90],
                  "Performance":[5,np.nan,2,np.nan,5]})
print(df)
print("\nNumber of nan values:",df.isna().sum())
df["Age"] = df["Age"].fillna(df["Age"].mean())
df["Marks"] = df["Marks"].fillna(df["Marks"].median())
df["Performance"] = df["Performance"].fillna(0)
print(df)

sns.barplot(x = "Name",y = "Marks",data = df)
plt.title("Marks of Student")
plt.show()

sns.lineplot(x = "Name",y = "Performance",data = df,marker = 'o')
plt.title("Performance Trend")
plt.show()
```
